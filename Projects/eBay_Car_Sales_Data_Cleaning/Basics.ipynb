## Cleaning

To begin cleaning:
- Columns with all or nearly all the same values are dropped
- Numeric data stored as text can be cleaned and converted

```python
autos.describe()
```

Columns that have little to no use:
- nrofpictures has only zero
- seller either private or company
- registration month does not tell much about the car
- offer type has few values that are necessary

```python
autos.drop(["nrofpictures","seller","registration_month","offer_type"],axis = 1)
```

For numeric data that is text (odometer and price):
- remove non-numeric data
- convert column to numeric
- for odometer, rename to odometer_km

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","").astype(float)
autos.rename({"odometer":"odometer_km"}, axis = 1, inplace = True)
```

```python
print(autos["price"].unique().shape)
print("\n")
print(autos["price"].describe())
print("\n")
print(autos["price"].value_counts())
print("\n")
print(autos["price"].value_counts().sort_index(ascending=False).head(20))
print("\n")
print(autos["price"].value_counts().sort_index().head(10))
```

```python
print(autos["odometer_km"].unique().shape)
print("\n")
print(autos["odometer_km"].describe())
print("\n")
print(autos["odometer_km"].value_counts())
```

Outliers to be removed:

For price, there are outliers that need to be removed above $20000 and below $200 as the prices above are single points and below are not values that cars normally are sold for.

```python
autos = autos[autos["price"].between(200,20000)]
```

In interest of the main car companies found in the data, the companies that make up 1% or more will be included in the analysis. 1% of the current data set is 43 cars and fewer than that is a reduction in statistical significance.

```python
mean_price = {}
for b in autos["brand"].value_counts(normalize=True, dropna=False).index[:17]:
    mean_price[b] = sum(autos.loc[autos["brand"]==b,"price"])/(autos[autos["brand"]==b].shape[0])
```

```python
mean_price
```

The most expensive cars on average are BMW, Mercedes_benz, and Audi. Renault and opel are the least expensive.

```python
mean_mileage = {}
for b in autos["brand"].value_counts(normalize=True, dropna=False).index[:17]:
    mean_mileage[b] = sum(autos.loc[autos["brand"]==b,"odometer_km"])/(autos[autos["brand"]==b].shape[0])
```

```python
mean_mileage
```

```python
bmp = pd.Series(mean_price)
bmm = pd.Series(mean_mileage)
```

```python
brand_pm = pd.DataFrame(bmp, columns = ['mean_price'])
brand_pm["mean_mileage"] = bmm
brand_pm
```

Audi has the highest number of kilometers on average and Smart has the fewest. Price does not seem to correlate with mileage.
